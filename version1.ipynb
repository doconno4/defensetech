```python
BASE_URL = "https://www.diu.mil/solutions/portfolio/catalog"
HEADERS = {"User-Agent": "Mozilla/5.0"}

def parse_catalog_page(html: str, page_num: int):
    """
    Extracts catalog entries using a robust heuristic:
    each result block shows:
      Project -> Focus Area (endswith 'Portfolio') -> Company -> Year -> Key Terms
    We identify Focus Area links (e.g., 'AI/ML Portfolio') and take:
      project = previous link text
      company = next link text
      year    = link after company
    """
    soup = BeautifulSoup(html, "lxml")
    text_links = [a.get_text(strip=True) for a in soup.find_all("a") if a.get_text(strip=True)]

    rows = []
    for i, t in enumerate(text_links):
        # Focus Area is almost always like "AI/ML Portfolio" (NOT the nav item "Portfolio")
        if t.endswith("Portfolio") and t != "Portfolio" and len(t) > len("Portfolio"):
            # guard rails for indices
            if i - 1 < 0 or i + 2 >= len(text_links):
                continue

            project = text_links[i - 1]
            focus_area = t
            company = text_links[i + 1]
            year = text_links[i + 2]

            # basic sanity checks: drop obvious nav garbage
            if project in {"Solutions", "Latest", "Contact", "Work With Us", "Who We Are"}:
                continue
            if company in {"Contact Us", "DIU for DoW Entities", "DIU for DoD Entities", "Portfolio"}:
                continue

            rows.append({
                "catalog_page": page_num,
                "project": project,
                "diu_focus_area": focus_area,
                "company_partner": company,
                "year_completed_raw": year,
                "source_url": f"{BASE_URL}?page={page_num}&q="
            })

    # De-dupe exact duplicates caused by repeated nav/footer links
    if rows:
        rows = pd.DataFrame(rows).drop_duplicates().to_dict("records")

    return rows


all_rows = []
page = 1

while True:
    url = f"{BASE_URL}?page={page}&q="
    r = requests.get(url, headers=HEADERS, timeout=30)

    if r.status_code != 200:
        print(f"Stop: status {r.status_code} on page {page}")
        break

    rows = parse_catalog_page(r.text, page)

    # If no rows, we’re likely past the last page
    if len(rows) == 0:
        print(f"Stop: no results parsed on page {page}")
        break

    all_rows.extend(rows)
    print(f"Page {page}: +{len(rows)} rows (total {len(all_rows)})")

    page += 1
    time.sleep(0.4)  # be polite


diu_catalog_entries = pd.DataFrame(all_rows)

# normalize year (some are like "11 Oct 2022" or "-")
diu_catalog_entries["year_completed"] = (
    diu_catalog_entries["year_completed_raw"]
    .astype(str)
    .str.extract(r"(\d{4})", expand=False)
)
diu_catalog_entries["year_completed"] = pd.to_numeric(diu_catalog_entries["year_completed"], errors="coerce")

diu_catalog_entries.head()
```

```python
USAS_BASE = "https://api.usaspending.gov/api/v2"
HEADERS = {
    "Content-Type": "application/json",
    "User-Agent": "defense-research"
}

FIELDS_USAS = [
    "Award ID",
    "Recipient Name",
    "Recipient UEI",
    "Award Amount",
    "Action Date",
    "Award Type",
    "Awarding Agency",
    "Funding Agency",
    "Description",
]
```

```python
AWARD_TYPE_GROUPS = {
    "contracts": ["A", "B", "C", "D"],
    "idvs": ["IDV_A", "IDV_B", "IDV_C", "IDV_D", "IDV_E"],
}
```

```python
def _payload_hash(payload: dict) -> str:
    s = json.dumps(payload, sort_keys=True).encode("utf-8")
    return hashlib.md5(s).hexdigest()
```

```python

def usas_pull_spending_by_award_dod(
    fy_list,
    award_type_codes,
    award_group_name,
    page_limit=100,
    max_pages=25,
    sleep_s=0.25
):
    """
    Pull DoD-only awards from USAspending for ONE award-type group.
    Saves each page to disk in RAW_USAS.
    """
    endpoint = f"{USAS_BASE}/search/spending_by_award/"

    base_payload = {
        "filters": {
            "agencies": [
                {
                    "type": "awarding",
                    "tier": "toptier",
                    "name": "Department of Defense"
                }
            ],
            "fy": fy_list,
            "award_type_codes": award_type_codes
        },
        "fields": FIELDS_USAS,
        "sort": "Award Amount",
        "order": "desc",
        "page": 1,
        "limit": page_limit,
    }

    all_rows = []
    page = 1
    has_next = True

    while has_next and page <= max_pages:
        payload = dict(base_payload)
        payload["page"] = page

        r = requests.post(endpoint, headers=HEADERS, json=payload)
        if r.status_code != 200:
            raise RuntimeError(
                f"USAspending error ({award_group_name}) {r.status_code}: {r.text[:500]}"
            )

        data = r.json()
        rows = data.get("results", [])
        meta = data.get("page_metadata", {}) or {}
        has_next = meta.get("hasNext", False)

        # Save each page immediately
        ph = _payload_hash(payload)
        outpath = (
            f"{RAW_USAS}/usas_award_raw_{award_group_name}_"
            f"fy{min(fy_list)}-{max(fy_list)}_p{page:04d}_{ph}.parquet"
        )
        pd.DataFrame(rows).to_parquet(outpath, index=False)

        all_rows.extend(rows)
        page += 1
        time.sleep(sleep_s)

    df_out = pd.DataFrame(all_rows)
    df_out["award_group"] = award_group_name
    return df_out
```

```python

fy_list = [2019, 2020, 2021, 2022, 2023, 2024, 2025]

dfs = []
for group_name, codes in AWARD_TYPE_GROUPS.items():
    print(f"Pulling {group_name}...")
    df = usas_pull_spending_by_award_dod(
        fy_list=fy_list,
        award_type_codes=codes,
        award_group_name=group_name,
        page_limit=100,
        max_pages=25,   # keep as safety cap first run
        sleep_s=0.25
    )
    dfs.append(df)

usas_award_raw = pd.concat(dfs, ignore_index=True)
usas_award_raw.shape
```

```python
import os, time, random
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timezone

HEADERS = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
}

PROJECT_DIR = "/content/defense_project"
PROCESSED_DIR = f"{PROJECT_DIR}/processed"
os.makedirs(PROCESSED_DIR, exist_ok=True)

def normalize_company_name(s):
    if not s:
        return None
    return " ".join(str(s).strip().split()).upper()

def get_html(url, timeout=(10, 90), max_retries=6, backoff=1.7):
    """
    Robust HTML fetch for SOSSEC:
    - longer read timeout (90s)
    - retries + exponential backoff + jitter
    """
    for attempt in range(1, max_retries + 1):
        try:
            r = requests.get(url, headers=HEADERS, timeout=timeout)
            if r.status_code == 200 and r.text:
                return r.text
            if r.status_code in (429, 500, 502, 503, 504):
                raise requests.exceptions.HTTPError(f"HTTP {r.status_code}")
            return None
        except (requests.exceptions.ReadTimeout,
                requests.exceptions.ConnectTimeout,
                requests.exceptions.ConnectionError,
                requests.exceptions.HTTPError) as e:
            sleep_s = (backoff ** (attempt - 1)) + random.uniform(0, 0.8)
            print(f"[retry {attempt}/{max_retries}] {type(e).__name__}: {e} | sleep {sleep_s:.1f}s")
            time.sleep(sleep_s)
    return None
```

```python
def scrape_sossec_member_directory_stopfast(max_pages=200, sleep_s=0.15, timeout=(10, 90)):
    base = "https://sossecinc.com/member-directory/"
    ingested_at = datetime.now(timezone.utc).isoformat()

    rows = []
    seen = set()
    no_new_streak = 0

    def page_url(p):
        return base if p == 1 else f"{base}page/{p}/"

    for p in range(1, max_pages + 1):
        url = page_url(p)
        html = get_html(url, timeout=timeout, max_retries=6)
        if html is None:
            print(f"[STOP] page {p}: fetch failed")
            break

        page_new = 0

        # Try read_html first (best if table exists)
        try:
            tables = pd.read_html(html)
        except ValueError:
            tables = []

        picked = None
        for t in tables:
            cols = [c.lower() for c in t.columns.astype(str)]
            if ("title" in cols and "email" in cols and "phone" in cols and "company memberships" in cols):
                picked = t
                break

        if picked is not None and len(picked) > 0:
            for _, r in picked.iterrows():
                company = str(r.get("Title", "")).strip()
                if not company or company.lower() == "nan":
                    continue

                company_std = normalize_company_name(company)
                if company_std in seen:
                    continue

                email = str(r.get("Email", "")).strip()
                phone = str(r.get("Phone", "")).strip()
                memberships = str(r.get("Company Memberships", "")).strip()
                mem_upper = memberships.upper() if memberships and memberships.lower() != "nan" else ""

                seen.add(company_std)
                page_new += 1
                rows.append({
                    "company_name_std": company_std,
                    "company_name_display": company,
                    "member_email": None if (not email or email.lower()=="nan") else email,
                    "member_phone": None if (not phone or phone.lower()=="nan") else phone,
                    "memberships_raw": None if (not memberships or memberships.lower()=="nan") else memberships,
                    "flag_sossec_member": int("SOSSEC" in mem_upper),
                    "flag_scec_member": int("SCEC" in mem_upper),
                    "source_system": "SOSSEC_MEMBER_DIRECTORY",
                    "source_url": url,
                    "ingested_at_utc": ingested_at,
                })
        else:
            # Fallback parse if read_html fails (keeps it resilient)
            soup = BeautifulSoup(html, "html.parser")
            text = soup.get_text("\n", strip=True)
            lines = [ln.strip() for ln in text.split("\n") if ln.strip()]
            for i in range(1, len(lines)):
                if "@" in lines[i] and "." in lines[i]:
                    company = lines[i-1]
                    company_std = normalize_company_name(company)
                    if not company_std or company_std in seen:
                        continue
                    seen.add(company_std)
                    page_new += 1
                    rows.append({
                        "company_name_std": company_std,
                        "company_name_display": company,
                        "member_email": lines[i],
                        "member_phone": lines[i+1] if i+1 < len(lines) else None,
                        "memberships_raw": None,
                        "flag_sossec_member": 0,
                        "flag_scec_member": 0,
                        "source_system": "SOSSEC_MEMBER_DIRECTORY",
                        "source_url": url,
                        "ingested_at_utc": ingested_at,
                    })

        print(f"[SOSSEC] page {p}: +{page_new} new | total_unique={len(seen)}")

        # Stop logic: if we’re not getting new companies, pagination isn’t adding value
        if page_new == 0:
            no_new_streak += 1
        else:
            no_new_streak = 0

        if no_new_streak >= 2:
            print("[STOP] no new companies across 2 pages → stopping (pagination likely repeats full list).")
            break

        time.sleep(sleep_s)

    df = pd.DataFrame(rows)
    if len(df) == 0:
        raise RuntimeError("SOSSEC member scrape returned 0 rows. Not saving.")
    df = df.drop_duplicates(subset=["company_name_std"]).sort_values("company_name_std").reset_index(drop=True)
    return df

df_sossec_members = scrape_sossec_member_directory_stopfast(max_pages=60)
print("unique companies:", df_sossec_members["company_name_std"].nunique())
display(df_sossec_members.head(25))

# Save
out = f"{PROCESSED_DIR}/company_universe_sossec_members.parquet"
df_sossec_members.to_parquet(out, index=False)
print("saved:", out)
```

```python
import os, re, time, random
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timezone

# Ensure PROJECT_DIR and PROCESSED_DIR are defined
PROJECT_DIR = "/content/defense_project"
PROCESSED_DIR = f"{PROJECT_DIR}/processed"
os.makedirs(PROCESSED_DIR, exist_ok=True)

HEADERS = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
}

def normalize_company_name(s):
    if not s:
        return None
    return " ".join(str(s).strip().split()).upper()

def get_html(url, timeout=(10, 90), max_retries=6, backoff=1.7):
    """
    Robust HTML fetch for SOSSEC:
    - longer read timeout (90s)
    - retries + exponential backoff + jitter
    """
    for attempt in range(1, max_retries + 1):
        try:
            r = requests.get(url, headers=HEADERS, timeout=timeout)
            if r.status_code == 200 and r.text:
                return r.text
            if r.status_code in (429, 500, 502, 503, 504):
                raise requests.exceptions.HTTPError(f"HTTP {r.status_code}")
            return None
        except (requests.exceptions.ReadTimeout,
                requests.exceptions.ConnectTimeout,
                requests.exceptions.ConnectionError,
                requests.exceptions.HTTPError) as e:
            sleep_s = (backoff ** (attempt - 1)) + random.uniform(0, 0.8)
            print(f"[retry {attempt}/{max_retries}] {type(e).__name__}: {e} | sleep {sleep_s:.1f}s")
            time.sleep(sleep_s)
    return None

def split_companies(blob):
    if not blob:
        return []
    blob = blob.replace(" & ", ", ").replace(" and ", ", ")
    parts = [p.strip(" .;:\n\t") for p in blob.split(",") if p.strip()]
    out, seen = [], set()
    for p in parts:
        k = p.upper()
        if k in seen:
            continue
        seen.add(k)
        out.append(p)
    return out

def parse_sossec_award_text(text):
    """
    Parses text to find award details and company names.
    Returns (awardees:list, project_title:str|None, award_date_str|None)
    """
    if not text: return ([], None, None)
    t = " ".join(text.split())

    # Attempt to extract project title (heuristic: often within <h2> or bold text)
    project_title = None
    title_match = re.search(r'\bproject\s+[^\n\r]+?\b(awarded|contract)\b', t, re.IGNORECASE)
    if title_match:
        project_title = title_match.group(0).strip()

    # Awardee extraction pattern (similar to NSTXL's 'awarded to')
    m = re.search(r"\b(?:awarded\s+to|contract\s+with|partners\s+with|selected\s+for)\b\s*([A-Za-z0-9,\s&\.-]+?)(?:\s+(?:on|for|in)\s+[0-9]{1,2}\s+[A-Za-z]+\s+[0-9]{4}|\s*\.|\s*\band\b|\s*\n|$)",
                  t, flags=re.IGNORECASE)

    awardees = []
    award_date = None

    if m:
        awardee_blob = m.group(1).strip()
        # Further clean the awardee_blob to remove trailing phrases like 'on date' or 'for project'
        awardee_blob = re.split(r'\s+(?:on|for|in)\s+[0-9]{1,2}\s+[A-Za-z]+\s+[0-9]{4}', awardee_blob, maxsplit=1, flags=re.IGNORECASE)[0]
        awardee_blob = re.split(r'\b(?:project|initiative|program|solution)\b', awardee_blob, maxsplit=1, flags=re.IGNORECASE)[0]
        awardee_blob = re.split(r'\b(?:for\s+a\s+contract|for\s+project|to\s+develop|to\s+provide)\b', awardee_blob, maxsplit=1, flags=re.IGNORECASE)[0]
        awardees = split_companies(awardee_blob)

        # Extract date if available (more generic pattern)
        date_match = re.search(r'([0-9]{1,2}\s+[A-Za-z]+\s+[0-9]{4})', t)
        if date_match:
            award_date = date_match.group(1)

    return (awardees, project_title, award_date)

def scrape_sossec_news_awards(max_pages=20, sleep_s=0.25):
    """
    Scrapes SOSSEC news/announcements for award details.
    """
    base_url = "https://sossecinc.com/news/"
    all_awards = []
    ingested_at = datetime.now(timezone.utc).isoformat()
    seen_awards = set() # To prevent duplicate award entries from different pages

    for page_num in range(1, max_pages + 1):
        url = f"{base_url}page/{page_num}/" if page_num > 1 else base_url
        html = get_html(url)
        if not html:
            print(f"[SOSSEC Awards] No content or failed fetch for page {page_num}. Stopping.")
            break

        soup = BeautifulSoup(html, "html.parser")
        articles = soup.find_all("article") # Assuming awards are in article tags

        if not articles:
            print(f"[SOSSEC Awards] No articles found on page {page_num}. Stopping.")
            break

        page_new_awards = 0
        for article in articles:
            link_tag = article.find("a", href=True)
            article_url = link_tag["href"] if link_tag else url
            article_title_tag = article.find(["h1", "h2", "h3"])
            article_title = article_title_tag.get_text(" ", strip=True) if article_title_tag else "No Title"
            article_text = article.get_text(" ", strip=True)

            awardees, project_title, award_date_str = parse_sossec_award_text(article_text)

            if awardees:
                for company_display_name in awardees:
                    company_std = normalize_company_name(company_display_name)
                    # Use a tuple of unique identifiers for an award to prevent duplicates
                    award_identifier = (company_std, project_title, award_date_str, article_url)
                    if award_identifier not in seen_awards:
                        all_awards.append({
                            "company_name_std": company_std,
                            "company_name_display": company_display_name,
                            "ota_manager": "SOSSEC",
                            "ota_vehicle": "SOSSEC OTA", # Placeholder, adjust if specific vehicle is found
                            "award_type": "OTA",
                            "award_signal": "awarded",
                            "project_title": project_title or article_title,
                            "award_date_text": award_date_str,
                            "raw_source_url": article_url,
                            "excerpt_text": article_text[:500], # Keep a snippet of the text
                            "ingested_at_utc": ingested_at,
                        })
                        seen_awards.add(award_identifier)
                        page_new_awards += 1
            else:
                print(f"[DEBUG] No awardees found in article: {article_url}\nText snippet: {article_text[:500]}...")

        print(f"[SOSSEC Awards] Page {page_num}: +{page_new_awards} new awards (Total: {len(all_awards)})")
        time.sleep(sleep_s + random.uniform(0.1, 0.5))

    return pd.DataFrame(all_awards)

# --- Run the scraper ---
```

```python
# ============================================================
# S2MARTS (NSTXL) — AWARD NOTICES (AWARDED OUTCOMES ONLY)
# Source pages: NSTXL year archives (2021–2024+)
# Saves:
#   /content/defense_project/processed/ota_s2marts_awarded.parquet
#   /content/defense_project/processed/company_universe_s2marts_awarded.parquet
# ============================================================

import os, re, time, random
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timezone

PROJECT_DIR = "/content/defense_project"
PROCESSED_DIR = f"{PROJECT_DIR}/processed"
os.makedirs(PROCESSED_DIR, exist_ok=True)

HEADERS = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Connection": "keep-alive",
}

def normalize_company_name(s):
    if not s: return None
    return " ".join(str(s).strip().split()).upper()

def get_html(url, timeout=(10, 90), max_retries=6, backoff=1.7):
    for attempt in range(1, max_retries + 1):
        try:
            r = requests.get(url, headers=HEADERS, timeout=timeout)
            if r.status_code == 200 and r.text:
                return r.text
            if r.status_code in (429, 500, 502, 503, 504):
                raise requests.exceptions.HTTPError(f"HTTP {r.status_code}")
            return None
        except (requests.exceptions.ReadTimeout,
                requests.exceptions.ConnectTimeout,
                requests.exceptions.ConnectionError,
                requests.exceptions.HTTPError) as e:
            sleep_s = (backoff ** (attempt - 1)) + random.uniform(0, 0.8)
            print(f"[retry {attempt}/{max_retries}] {type(e).__name__}: {e} | sleep {sleep_s:.1f}s")
            time.sleep(sleep_s)
    return None

def split_companies(blob):
    if not blob: return []
    blob = blob.replace(" & ", ", ").replace(" and ", ", ")
    parts = [p.strip(" .;:\n\t") for p in blob.split(",") if p.strip()]
    out, seen = [], set()
    for p in parts:
        k = p.upper()
        if k in seen: continue
        seen.add(k); out.append(p)
    return out

def parse_award(text):
    """
    Handles NSTXL award notice patterns like:
      - "This opportunity was awarded to X on 08 March 2024."
      - "This opportunity was awarded to X, Y, and Z on 20 October 2022."
      - Sometimes no date: "This opportunity was awarded to Qualcomm Technologies and Microsoft"
    Returns (awardees:list, award_date_str|None)
    """
    if not text: return ([], None)
    t = " ".join(text.split())

    m = re.search(r"\bawarded to\b\s*(.+?)(?:\s+on\s+([0-9]{1,2}\s+[A-Za-z]+\s+[0-9]{4}))?(?:\.|$)",
                  t, flags=re.IGNORECASE)
    if not m:
        return ([], None)

    awardee_blob = m.group(1).strip()
    award_date = (m.group(2).strip() if m.group(2) else None)

    # cut off after common boilerplate markers
    awardee_blob = re.split(r"\bSolutions Due\b|\bQuestions Due\b|\bRFS Release\b|\bQ&A\b|\*\s\*\s\*",
                            awardee_blob, maxsplit=1, flags=re.IGNORECASE)[0].strip()

    return (split_companies(awardee_blob), award_date)

def parse_project_no(text):
    if not text: return None
    # "S²MARTS Project No. 23-08" or "Project No. 20-10A"
    m = re.search(r"\bProject\s+No\.\s*([0-9]{2}\-[0-9]{2}[A-Z]?(?:\s*[A-Z])?)\b", text, flags=re.IGNORECASE)
    return m.group(1).replace(" ", "") if m else None

def scrape_nstxl_year_archive(year):
    url = f"https://nstxl.org/categories/{year}/"
    html = get_html(url)
    if html is None:
        return []
    soup = BeautifulSoup(html, "html.parser")

    rows = []
    for art in soup.find_all("article"):
        title_tag = art.find(["h2","h3"])
        a = art.find("a", href=True)
        title = title_tag.get_text(" ", strip=True) if title_tag else None
        link = a["href"] if a else None
        excerpt = art.get_text(" ", strip=True)

        awardees, award_date = parse_award(excerpt)
        if not awardees:
            continue

        proj_no = parse_project_no(excerpt) or parse_project_no(title)
        rows.append((title, link, excerpt, proj_no, award_date, awardees))
    return rows

# ---- RUN across years (edit range if you want more/less) ----
years = list(range(2020, 2026))  # includes older + current; adjust anytime
ingested_at = datetime.now(timezone.utc).isoformat()
out_rows = []

for y in years:
    hits = scrape_nstxl_year_archive(y)
    print(f"[NSTXL] {y}: award posts found {len(hits)}")
    for title, link, excerpt, proj_no, award_date, awardees in hits:
        for comp in awardees:
            out_rows.append({
                "company_name_std": normalize_company_name(comp),
                "company_name_display": comp,
                "ota_manager": "NSTXL",
                "ota_vehicle": "S2MARTS",
                "award_type": "OTA",
                "award_signal": "awarded",
                "project_number": proj_no,
                "project_title": title,
                "award_date_text": award_date,
                "raw_source_url": link,
                "excerpt_text": excerpt[:600],
                "ingested_at_utc": ingested_at,
            })

ota_s2marts_awarded = pd.DataFrame(out_rows, columns=[
    "company_name_std","company_name_display","ota_manager","ota_vehicle","award_type","award_signal",
    "project_number","project_title","award_date_text","raw_source_url","excerpt_text","ingested_at_utc"
])

ota_s2marts_awarded = ota_s2marts_awarded.dropna(subset=["company_name_std","raw_source_url"])
ota_s2marts_awarded = ota_s2marts_awarded.drop_duplicates(subset=["company_name_std","raw_source_url","project_number"])

print("S2MARTS awarded rows:", len(ota_s2marts_awarded))
display(ota_s2marts_awarded.head(25))

if len(ota_s2marts_awarded) == 0:
    raise RuntimeError("NSTXL S2MARTS award scrape returned 0 rows. Not saving.")

# Save award table
out_aw = f"{PROCESSED_DIR}/ota_s2marts_awarded.parquet"
ota_s2marts_awarded.to_parquet(out_aw, index=False)
print("saved:", out_aw)

# Save company universe from awarded outcomes only
company_universe_s2marts_awarded = (
    ota_s2marts_awarded[["company_name_std"]]
    .drop_duplicates()
    .sort_values("company_name_std")
    .reset_index(drop=True)
)

out_univ = f"{PROCESSED_DIR}/company_universe_s2marts_awarded.parquet"
company_universe_s2marts_awarded.to_parquet(out_univ, index=False)
print("saved:", out_univ, "| unique companies:", len(company_universe_s2marts_awarded))
```

```python
import re, time, random, requests, pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime, timezone

HEADERS = {"User-Agent": "Mozilla/5.0", "Accept-Language": "en-US,en;q=0.9"}

def get_html(url, timeout=(10, 90), retries=5, backoff=1.6):
    for a in range(retries):
        try:
            r = requests.get(url, headers=HEADERS, timeout=timeout)
            if r.status_code == 200 and r.text:
                return r.text
            if r.status_code in (429, 500, 502, 503, 504):
                raise Exception(f"HTTP {r.status_code}")
            return None
        except Exception:
            time.sleep((backoff ** a) + random.uniform(0, 0.6))
    return None

def normalize(s):
    return " ".join(str(s).strip().split()).upper() if s else None

def split_companies(blob):
    if not blob: return []
    blob = blob.replace(" & ", ", ").replace(" and ", ", ")
    parts = [p.strip(" .;:\n\t") for p in blob.split(",") if p.strip()]
    out, seen = [], set()
    for p in parts:
        k = p.upper()
        if k in seen:
            continue
        seen.add(k)
        out.append(p)
    return out

def parse_awarded_to(text):
    if not text: return []
    t = " ".join(text.split())
    m = re.search(r"\bawarded to\b\s*(.+?)(?:\s+on\s+[0-9]{1,2}\s+[A-Za-z]+\s+[0-9]{4})?(?:\.|$)", t, flags=re.IGNORECASE)
    if not m:
        return []
    tail = m.group(1).strip()
    tail = re.split(r"\bSolutions Due\b|\bQuestions Due\b|\bRFS Release\b|\bQ&A\b", tail, maxsplit=1, flags=re.IGNORECASE)[0].strip()
    return split_companies(tail)

def crawl_index_pages(start_urls, max_pages_each=50):
    """Collect post URLs from index/category pages with pagination."""
    post_urls = set()
    for base in start_urls:
        for p in range(1, max_pages_each + 1):
            url = base if p == 1 else (base.rstrip("/") + f"/page/{p}/")
            html = get_html(url)
            if not html:
                break
            soup = BeautifulSoup(html, "html.parser")
            arts = soup.find_all("article")
            if not arts:
                break
            before = len(post_urls)
            for art in arts:
                a = art.find("a", href=True)
                if a and a["href"]:
                    post_urls.add(a["href"])
            if len(post_urls) == before:
                break
            time.sleep(0.15)
    return sorted(post_urls)

# Broad coverage: year archives + relevant categories
start_urls = [
    "https://nstxl.org/categories/2019/",
    "https://nstxl.org/categories/2020/",
    "https://nstxl.org/categories/2021/",
    "https://nstxl.org/categories/2022/",
    "https://nstxl.org/categories/2023/",
    "https://nstxl.org/categories/2024/",
    "https://nstxl.org/categories/2025/",
    "https://nstxl.org/categories/s2marts-research/",
    # if they have an 'awarded' category on NSTXL (many do), include it if it exists:
    "https://nstxl.org/categories/awarded/",
]

post_urls = crawl_index_pages(start_urls, max_pages_each=40)
print("Collected post URLs:", len(post_urls))

rows = []
ingested_at = datetime.now(timezone.utc).isoformat()

for i, u in enumerate(post_urls, 1):
    html = get_html(u)
    if not html:
        continue
    soup = BeautifulSoup(html, "html.parser")
    text = soup.get_text(" ", strip=True)

    awardees = parse_awarded_to(text)
    if not awardees:
        continue

    title = soup.find("h1")
    title = title.get_text(" ", strip=True) if title else None

    for c in awardees:
        rows.append({
            "company_name_std": normalize(c),
            "company_name_display": c,
            "award_type": "OTA",
            "ota_manager": "NSTXL",
            "ota_vehicle": "S2MARTS",
            "award_signal": "awarded",
            "project_title": title,
            "raw_source_url": u,
            "ingested_at_utc": ingested_at,
        })

    if i % 50 == 0:
        print("processed posts:", i, "| award rows so far:", len(rows))

aw = pd.DataFrame(rows)

print("\nAward rows:", len(aw))
if len(aw) > 0:
    print("Unique companies:", aw["company_name_std"].nunique())
    display(aw.head(25))
else:
    print("No awardees found yet. That means the index pages didn’t include the award posts we need, not that awards don’t exist.")
```

```python
# ----------------------------
# HigherGov (company-centric pull)
# Fill this cell only (Cell 59). Uses existing: RAW, vc_core, normalize_company_name()
# ----------------------------
import os, json, time, hashlib, requests
import pandas as pd

RAW_HG = f"{RAW}/highergov"
RAW_HG_AWARDEE = f"{RAW_HG}/awardee"
RAW_HG_PULLS = f"{RAW_HG}/pulls"   # per-company cached pulls (contracts/vehicles/grants)
for d in [RAW_HG, RAW_HG_AWARDEE, RAW_HG_PULLS]:
    os.makedirs(d, exist_ok=True)

HG_BASE = "https://www.highergov.com/api-external"
HG_API_KEY = os.environ.get("HG_API_KEY")
if not HG_API_KEY:
    raise ValueError("Missing HG_API_KEY. In Colab: Secrets -> add HG_API_KEY, then restart runtime.")

HG_HEADERS = {"User-Agent": "defense-startup-research"}

def _md5(s: str) -> str:
    return hashlib.md5(s.encode("utf-8")).hexdigest()

def hg_get_json(endpoint: str, params: dict, cache_path: str | None = None, timeout_s: int = 60):
    """
    Simple GET with optional caching to disk.
    """
    url = f"{HG_BASE}/{endpoint.strip('/')}/"
    p = dict(params)
    p["api_key"] = HG_API_KEY

    if cache_path and os.path.exists(cache_path):
        with open(cache_path, "r") as f:
            return json.load(f)

    r = requests.get(url, params=p, headers=HG_HEADERS, timeout=timeout_s)
    r.raise_for_status()
    data = r.json()

    if cache_path:
        with open(cache_path, "w") as f:
            json.dump(data, f)

    return data

def hg_get_paginated(endpoint: str, params: dict, cache_dir: str, page_size: int = 250, max_pages: int = 25, sleep_s: float = 0.10):
    """
    HigherGov pagination pattern: page_number/page_size.
    Caches each page so re-runs don’t burn trial credits.
    """
    os.makedirs(cache_dir, exist_ok=True)
    out = []

    for page in range(1, max_pages + 1):
        p = dict(params)
        p["api_key"] = HG_API_KEY
        p["page_number"] = page
        p["page_size"] = page_size

        cache_path = f"{cache_dir}/{endpoint}_p{page:04d}_{_md5(json.dumps(p, sort_keys=True))}.json"
        if os.path.exists(cache_path):
            with open(cache_path, "r") as f:
                data = json.load(f)
        else:
            url = f"{HG_BASE}/{endpoint.strip('/')}/"
            r = requests.get(url, params=p, headers=HG_HEADERS, timeout=60)
            r.raise_for_status()
            data = r.json()
            with open(cache_path, "w") as f:
                json.dump(data, f)
            time.sleep(sleep_s)

        results = data.get("results", [])
        if not results:
            break
        out.extend(results)

        total_pages = data.get("total_pages")
        if isinstance(total_pages, int) and page >= total_pages:
            break

    return pd.json_normalize(out)

# ----------------------------
# 1) Company universe (from existing vc_core)
# ----------------------------
hg_company_universe = (
    vc_core[["company_name_std"]]
      .dropna()
      .drop_duplicates()
      .reset_index(drop=True)
)

# ----------------------------
# 2) Awardee resolution (optional but recommended)
#    Converts company -> (awardee_id / UEI / CAGE) if HigherGov returns them
# ----------------------------
def hg_awardee_lookup(company_name_std: str, top_n: int = 5):
    """
    Best-effort: HigherGov Awardee endpoint search param names can differ.
    We try several common keys; first hit wins.
    """
    name = str(company_name_std).strip()
    cache_path = f"{RAW_HG_AWARDEE}/awardee_lookup_{_md5(name)}.json"

    param_candidates = [
        {"keyword": name},
        {"q": name},
        {"search": name},
        {"name": name},
    ]

    for base in param_candidates:
        params = {**base, "page_number": 1, "page_size": top_n}
        data = hg_get_json("awardee", params=params, cache_path=cache_path)
        results = data.get("results", [])
        if results:
            return results
    return []

awardee_rows = []
for co in hg_company_universe["company_name_std"].tolist():
    results = hg_awardee_lookup(co, top_n=5)
    top = results[0] if results else {}
    awardee_rows.append({
        "company_name_std": co,
        "hg_awardee_name": top.get("name") or top.get("awardee_name"),
        "hg_awardee_id": top.get("id") or top.get("awardee_id"),
        "hg_uei": top.get("uei"),
        "hg_cage": top.get("cage"),
        "hg_awardee_hit": bool(top),
    })

hg_awardee_map = pd.DataFrame(awardee_rows)

# Join mapping into vc_core (safe left join; doesn't break downstream)
vc_core = vc_core.merge(hg_awardee_map, on="company_name_std", how="left")
print("hg_awardee_hit rate:", vc_core["hg_awardee_hit"].mean())
```
